```python
from PyPDF2 import PdfReader

reader = PdfReader("questionPool.pdf")

text = ""  
for page in reader.pages:
    text += page.extractText() + "\n"

f = open("demofile2.txt", "w")
f.write(text)
f.close()
!head demofile2.txt
```

```python
import re

read = open("demofile2.txt", "r")
result = open("format.txt", "w")

prevLine = ""
busy = False
for line in read:
  line = line.strip()
  hasDigit = line[:1].isdigit()
  option = re.findall("^[a-f]\. (\w.+)", line)
 
  if hasDigit:
    prevLine = line
    busy = True
  elif len(line) != 0:
    if not option:
      prevLine += " " + line
    else:
      if busy:
        result.write(prevLine + "\n")
        busy = False
      result.write(line + "\n")

result.close()
read.close()

!tail format.txt
```

### CSV works



```python
import re
mcqs = open("demofile2.txt", "r")

csv = open("results.csv", "w")
csv.write("No,Question,Answer1,AnswerCorrect1,Answer2,AnswerCorrect2,Answer3,AnswerCorrect3,Answer4,AnswerCorrect4,Answer5,AnswerCorrect5,Answer6,AnswerCorrect6,Answer7,AnswerCorrect7,Answer8,AnswerCorrect8,Answer9,AnswerCorrect9,Answer10,AnswerCorrect10,Hint,Explanation,Category\n")

string = ""
for line in mcqs:
  line = line.strip()
  hasDigit = line[:1].isdigit()
  option = re.findall("^[a-f]\. (\w.+)", line)

  if hasDigit:
    quesPos = line.find(".")
    string += line[:quesPos] + ","
    string += "\"" + line[quesPos+2:] + "\""
  elif option:
    string += "," + option[0] + ","
  elif len(line) != 0:
    string += "\"" + line + "\""
  else:
    csv.write(string + "\n")
    string = ""

csv.close()
```
