## PDF to .txt

```python
from PyPDF2 import PdfReader

reader = PdfReader("questionPool.pdf")

text = ""  
for page in reader.pages:
    text += page.extractText() + "\n"

f = open("demofile2.txt", "w")
f.write(text)
f.close()
!tail -n 25 demofile2.txt
```

### Format demofile2.txt

```python
import re

read = open("demofile2.txt", "r")
result = open("format.txt", "w")

question = ""
hasQuestion = False
for line in read:
  line = line.strip()
  hasDigit = line[:1].isdigit()
  option = re.findall("^[a-f]\. (\w.+)", line)
 
  if hasDigit:
    question = line
    hasQuestion = True
  elif len(line) != 0:
    if not option:
      question += " " + line
    else:
      if hasQuestion:
        result.write("\n" + question + "\n")
        hasQuestion = False
      result.write(line + "\n")

result.close()
read.close()

!tail format.txt
```

### Get Answers

```python
import pdfplumber
f = open("answers.txt", "w")
with pdfplumber.open('questionPool.pdf') as pdf: 
    for page in pdf.pages:
        try:
            clean_text = page.filter(lambda obj: obj["object_type"] == "char" and "Bold" in obj["fontname"])
            f.write(clean_text.extract_text())
        except IndexError:
            print("")
            break
f.close()
!head answers.txt
```

```python
import os

f = open("answers.txt", "r")
edited = open("answers.md", "w")
answers = []

for line in f:
  if len(line.strip()) != 0:
    edited.write(line.strip() + "\n")
    answers.append(line.strip())

os.remove("answers.txt") 
!head answers.md
```

## TXT to CSV

### CSV v1



```python
import re
mcqs = open("demofile2.txt", "r")

csv = open("results.csv", "w")
csv.write("No,Question,Answer1,AnswerCorrect1,Answer2,AnswerCorrect2,Answer3,AnswerCorrect3,Answer4,AnswerCorrect4,Answer5,AnswerCorrect5,Answer6,AnswerCorrect6,Answer7,AnswerCorrect7,Answer8,AnswerCorrect8,Answer9,AnswerCorrect9,Answer10,AnswerCorrect10,Hint,Explanation,Category\n")

string = ""
for line in mcqs:
  line = line.strip()
  hasDigit = line[:1].isdigit()
  option = re.findall("^[a-f]\. (\w.+)", line)

  if hasDigit:
    quesPos = line.find(".")
    string += line[:quesPos] + ","
    string += "\"" + line[quesPos+2:] + "\""
  elif option:
    string += "," + option[0] + ","
  elif len(line) != 0:
    string += "\"" + line + "\""
  else:
    csv.write(string + "\n")
    string = ""

csv.close()
```
