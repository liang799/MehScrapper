```python
from PyPDF2 import PdfReader

reader = PdfReader("questionPool.pdf")

text = ""  
for page in reader.pages:
    text += page.extract_text() + "\n"

f = open("demofile2.txt", "w")
f.write(text)
f.close()
```

### CSV works



```python
mcqs = open("demofile2.txt", "r")

csv = open("results.csv", "w")
csv.write("No,Question,Answer1,AnswerCorrect1,Answer2,AnswerCorrect2,Answer3,AnswerCorrect3,Answer4,AnswerCorrect4,Answer5,AnswerCorrect5,Answer6,AnswerCorrect6,Answer7,AnswerCorrect7,Answer8,AnswerCorrect8,Answer9,AnswerCorrect9,Answer10,AnswerCorrect10,Hint,Explanation,Category\n")

string = ""
workbook = xlsxwriter.Workbook('results.xlsx')
for line in mcqs:
  line = line.strip()
  hasDigit = line[:1].isdigit()
  option1 = line.find("a.")
  option2 = line.find("b.")
  option3 = line.find("c.")
  option4 = line.find("d.")

  if hasDigit:
    quesPos = line.find(".")
    string += line[:quesPos] + ","
    string += "\"" + line[quesPos+2:] + "\""
  elif option1 == 0 or option2 == 0 or option3 == 0 or option4 == 0:
    string += "," + line[3:] + ","
  elif len(line) != 0:
    string += "\"" + line + "\""
  else:
    csv.write(string + "\n")
    string = ""

csv.close()
```
